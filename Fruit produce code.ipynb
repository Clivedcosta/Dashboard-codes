```python
# Species data dictionary
species_data = {
    "Teak": {"scientific_name": "Tectona grandis", "timber_per_tree": (650, 700), "average_height": "40m", "average_price": (3500,)},
    "Sheesham": {"scientific_name": "Dalbergia sissoo", "timber_per_tree": (750, 850), "average_height": "35-40m", "average_price": (2500,)},
    "Gamhar": {"scientific_name": "Gmelina arborea", "timber_per_tree": (400, 550), "average_height": "25-30m", "average_price": (1500,)},
    "Lal chandan": {"scientific_name": "Pterocarpus santilus", "timber_per_tree": (600, 650), "average_height": "30m", "average_price": (9500,)},
    "Semal": {"scientific_name": "Bombax ceiba", "timber_per_tree": (350, 400), "average_height": "20m", "average_price": (950,)},
    "Acacia": {"scientific_name": "Acacia auriculiformis", "timber_per_tree": (500, 600), "average_height": "15-30m", "average_price": (1000,)}
}

# Function to calculate total timber produced for a species
def calculate_timber(species, number_of_trees):
    # Check if the species exists in the data
    if species in species_data:
        timber_range = species_data[species]["timber_per_tree"]
        average_timber = (timber_range[0] + timber_range[1]) / 2  # Taking the average of the timber range
        total_timber_kg = average_timber * number_of_trees
        return total_timber_kg
    else:
        return "Species not found. Please check the species name."

# Function to calculate the price value for the timber produced by a species
def calculate_price(species, total_timber_kg):
    if species in species_data:
        price_per_cubic_meter = species_data[species]["average_price"][0]
        # Assume 1 cubic meter of timber weighs 1000 kg (this can be adjusted based on timber density)
        total_timber_cubic_meters = total_timber_kg / 1000
        total_price = total_timber_cubic_meters * price_per_cubic_meter
        return total_price
    else:
        return "Species not found. Please check the species name."

# Function to handle multiple species input and calculate total timber and total price
def calculate_total_timber_and_value():
    total_timber_kg = 0  # Total timber in kilograms
    total_value = 0  # Total price value in Rupees

    while True:
        species_input = input("Enter the species name (or type 'done' to finish): ").strip()
        if species_input.lower() == 'done':
            break

        number_of_trees_input = int(input(f"Enter the number of trees for {species_input}: ").strip())

        # Calculate total timber produced
        total_timber_for_species = calculate_timber(species_input, number_of_trees_input)
        if isinstance(total_timber_for_species, str):  # Check if the species was not found
            print(total_timber_for_species)
        else:
            total_timber_kg += total_timber_for_species
            # Calculate total value based on the price per cubic meter
            total_price_for_species = calculate_price(species_input, total_timber_for_species)
            total_value += total_price_for_species

            print(f"The total timber produced from {number_of_trees_input} {species_input} trees is {total_timber_for_species:.2f} kg.")
            print(f"The total value generated from {species_input} is {total_price_for_species:.2f} Rupees.\n")

    total_timber_ton = total_timber_kg / 1000  # Convert total timber to tons
    print(f"\nThe total timber produced from all species is {total_timber_ton:.2f} tons.")
    print(f"The total value generated from all species is {total_value:.2f} Rupees.")

# Example usage
calculate_total_timber_and_value()


```

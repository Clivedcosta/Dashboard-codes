```python
#This code it to calculate the area of a plantation based on the number of trees planted per hectare and the total number of trees planted.
def calculate_plantation_area(total_trees, trees_per_hectare):
    """Calculates the area of a plantation in hectares.

    Args:
        total_trees: The total number of trees planted.
        trees_per_hectare: The number of trees planted per hectare.

    Returns:
        The area of the plantation in hectares.
        Returns an error message if trees_per_hectare is invalid.
    """

    if trees_per_hectare <= 0:
      return "Error: Trees per hectare must be greater than zero."
    
    area_hectares = total_trees / trees_per_hectare
    return area_hectares


def main():
    """Gets user input and calculates plantation area."""
    try:
        total_trees = int(input("Enter the total number of trees: "))
        trees_per_hectare = float(input("Enter the number of trees planted per hectare: ")) # Allows for decimal input

        if total_trees < 0:
          raise ValueError("Total trees must be non-negative.")

        area_result = calculate_plantation_area(total_trees, trees_per_hectare)

        if isinstance(area_result, str): # Check if it's an error message
          print(area_result)
          return # Exit the function if there's an error

        area_hectares = area_result

        print(f"The approximate area of the plantation is {area_hectares:.2f} hectares.")

    except ValueError as e:
        print(f"Invalid input: {e}")  # Handles incorrect input (non-numbers, negative trees)
    except Exception as e: # Catches other potential errors
        print(f"An error occurred: {e}")


if __name__ == "__main__":
    main()
```
